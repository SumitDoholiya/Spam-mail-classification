```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#1
df = pd.read_csv('/content/mail_data (1).csv')
#2
df
```

```python
#3.
# replace null values with null string
df = df.fillna('null')
df
```

```python
#4.first 5 rows of the dataset
df.head(5)
```

```python
df.shape
```

```python
df.info
```

TASK 2 [ DATA PRE PROCESSING ] 1 D.S


```python
#3. (a) Assign 'spam' as 0 and 'ham' as 1
df.loc[df['Category'] == "spam", "Category"] = 0
df.loc[df['Category'] == "ham", "Category"] = 1
```

SEPARATING THE DATA AS TESXTS AND LABELS

```python
x = df['Message']
y = df['Category']

print(x)
```

```python
print(y)
```

```python

# Split data into training and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=3)

# Display the result
print(x.shape)
print(x_train.shape)
```

TRANSFORM THE TEXT DATA YO FEATURE VECTORS THAT CAN BE USED AS INPUT TO LOGISTIC REGRESSION

```python
from sklearn.preprocessing import StandardScaler,LabelEncoder


# Initialize the tfidfVectorizer
vectorizer = TfidfVectorizer(min_df = 1, stop_words='english', lowercase = True)

# Fit and transform the training text data
x_train_vec = vectorizer.fit_transform(x_train)

# Transform the test text data using the same vectorizer
x_test_vec = vectorizer.transform(x_test)

print(x_train_vec)

# Display the shape of the resulting vectors to verify
print("Training Data Vector Shape:", x_train_vec.shape)
print("Test Data Vector Shape:", x_test_vec.shape)
```

CONVERTING y_train AND y_test TO INTEGERS

```python
y_train = y_train.astype('int')
y_test = y_test.astype('int')

print(y_train)
```

[2] Training logistic regression model with training data.

TRAINING THE MODEL USING LOGISTIC REGRESSION

```python
log_reg = LogisticRegression()

log_reg.fit(x_train_vec, y_train)
```

PREDICTION ON TRAINING DATA & ACCURACY

```python
prediction_training_data = log_reg.predict(x_train_vec)
print(prediction_training_data)
accuracy_train = accuracy_score(y_train,prediction_training_data)
print(f"{accuracy_train * 100:.2f}%")
```

PREDICTION ON TEST DATA

```python
prediction_training_data = log_reg.predict(x_test_vec)
print(prediction_training_data)
accuracy_train = accuracy_score(y_test,prediction_training_data)
print(f"{accuracy_train * 100:.2f}%")
```

SAVING THE TRAINED MODEL

```python
import pickle
```

```python
filename = 'spam_ham_model_sav'
pickle.dump(log_reg, open(filename, 'wb'))
```

```python
# loading the save model

loaded_model = pickle.load(open('spam_ham_model_sav', 'rb'))
```

```python
# Save the Vectorizer
pickle.dump(vectorizer, open('vectorizer.pkl', 'wb'))
```

MAKING PREDICTOR SYSTEM

```python
# Sample Input Mail Prediction
input_mail = ["Free entry in 2 a wkly comp to win FA Cup final tkts 21st May 2005. Text FA to 87121 to receive entry question(std txt rate)T&C's apply 08452810075over18's"]

# Convert text to feature vectors
input_data_features = vectorizer.transform(input_mail)

# Making prediction
prediction = loaded_model.predict(input_data_features)
print("Prediction:", prediction)

if prediction[0] == 1:
    print('Ham mail')
else:
    print('Spam mail')
```
